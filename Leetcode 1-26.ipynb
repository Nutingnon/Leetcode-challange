```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        h = dict()
        for i,j in enumerate(nums):
            rest = target - j
            if j not in h:
                h[rest] = i
            else:
                return [h[j],i]
```

# 3. Longest Substring Without Repeating Characters

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        ls = ''
        l = 0
        for c in s:
            if c in ls:
                if len(ls) > l:
                    l = len(ls)
                ls = ls[ls.index(c)+1:]
            ls+=c
        if len(ls) > l:
            l = len(ls)
        return l
 
```

# Zigzag conversion

```python
def convert(s: str, numRows: int) -> str:
    if numRows <= 1:
        return s
    rows = [""] * numRows
    idx = 0
    direction = -1
    for i in range(len(s)):
        rows[idx] = rows[idx] + s[i]
        if i % (numRows-1) == 0:
            direction *= -1
        idx += direction
    return "".join(rows)
```

# 11 Container with most water

```python
def maxArea(self, height):
    i, j = 0, len(height) - 1
    water = 0
    while i < j:
        water = max(water, (j - i) * min(height[i], height[j]))
        if height[i] < height[j]:
            i += 1
        else:
            j -= 1
    return water
```

# Integer to Roman

```python
def intToRoman(self, num):
    """
    :type num: int
    :rtype: str
    """
    num = int(num)
    M = ["",'M','MM','MMM']
    C = ["",'C','CC','CCC','CD','D','DC','DCC','DCCC','CM']
    X = ['','X','XX','XXX','XL','L','LX','LXX','LXXX','XC']
    I = ['','I','II','III','IV','V','VI','VII','VIII','IX']
    return M[num//1000] + C[(num%1000)//100] + X[(num%100)//10] + I[num%10]
```

```python
def romanToInt(self, s):
    """
    :type s: str
    :rtype: int
    """
    dict1 = {'I':1,  'V':5, 'X':10,'L':50,'C':100,'D':500,'M':1000}
    int1 = 0
    i = 0
    while i <= len(s)-1:
        if i < len(s)-1:
            if dict1[s[i]] < dict1[s[i+1]]:
                int1+= dict1[s[i+1]] - dict1[s[i]]
                i += 2
            else:
                int1 += dict1[s[i]]
                i +=1
        else:
            int1 += dict1[s[i]]
            i += 1

    return int1

# I             1
# V             5
# X             10
# L             50
# C             100
# D             500
# M             1000
```

# 14 longest common prefix

```python
def longestCommonPrefix(self, strs):
    """
    :type strs: List[str]
    :rtype: str
    """
    if not strs: return ''
            #since list of string will be sorted and retrieved min max by alphebetic order
    s1 = min(strs)
    s2 = max(strs)
    for i, c in enumerate(s1):
        if c != s2[i]:
            return s1[:i] #stop until hit the split index
    return s1
```

```python
def threeSum(nums):
    outcome = []
    sl = sorted(nums)
    for n,m in enumerate(sl[:-2]):
        left = n+1
        right = len(nums) - 1
        while left < right:
            total = m + sl[left] + sl[right]
            if total < 0:
                left += 1
            elif total >0:
                right -=1
            else:
                outcome.append([m,sl[left],sl[right]])
                left +=1
                right -= 1
    return outcome
            
```

```python
def threeSum(nums):
    """
    :type nums: List[int]
    :rtype: List[List[int]]
    """
    instances = {}
    for n in nums:
        if n in instances:
            instances[n] += 1
        else:
            instances[n] = 1
    values = []
    result = []
    for n, count in sorted(instances.items()):
        values.append(n)
        if n == 0 and count >= 3:
            result.append([0,0,0])
        elif n != 0 and count >= 2:
            third = -2*n
            if third in instances:
                if n < third:
                    result.append([n,n,third])
                else:
                    result.append([third,n,n])
                    
    #any sums involving duplicate values were handled above
    nvalues = len(values)
    while nvalues >= 3:
        floor = -(values[nvalues-1] + values[nvalues-2]) 
        ceiling = -(values[0] + values[1])
        if floor > ceiling:
            return []
        iLeft = nvalues
        iRight = -1
        for i in range(nvalues):
            if values[i] >= floor:
                iLeft = i
                break
        for i in range(nvalues-1, -1, -1):
            if values[i] <= ceiling:
                iRight = i
                break
        if iLeft == 0 and iRight == nvalues - 1:
            break
        values = values[iLeft:iRight+1]
        nvalues = len(values)
    if nvalues < 3:
        return result
        
    for i in range(nvalues-2):
        v1 = values[i]
        if v1 >= 0:
            break
        for j in range(i+1,nvalues-1):
            v2 = values[j]
            v3 = -(v1 + v2)
            if v3 <= v2:
                break
            if v3 in instances:
                result.append([v1,v2,v3])
    return result
```

```python
def threeSum_C(nums, target):
    """
    :type nums: List[int]
    :rtype: List[List[int]]
    """
    nums.sort()
    min_ = abs(sum(nums[:3]) - target)
    record = sum(nums[:3])
    len_nums = len(nums)
    for i in range(len(nums)-2):
        if i >=1  and nums[i] == nums[i-1]:
                continue
        left, right = i+1, len_nums-1
        while left < right:
            s = nums[i]+nums[left] + nums[right]
            ds = abs(s - target)
            if ds == 0:
                return 0
            if ds >= min_ and s - target<0:
                left+= 1
            elif ds >= min_ and s - target>0:
                right -= 1
            elif ds < min_  and s - target >0:
                record = nums[i]+nums[left] + nums[right]
                min_ = ds
                right -= 1
            elif ds<min_ and s-target <0:
                min_ = ds
                record = nums[i]+nums[left] + nums[right]
                left += 1
                
    return record
```

# 17. Letter combinations of a phone number

```python

def letterCombinations(digits):
    """
    :type digits: str
    :rtype: List[str]
    """
    d = ['','','abc','def','ghi','jkl','mno','pqrs','tuv','wxyz']
    if len(digits) == 0:
        return []
    elif len(digits) == 1:
        return list(d[int(digits[0])])
    else:
        length = len(digits)
        output1 = ['']
        for i in digits:
            output_temp2 = []
            for j in output1:
                for l in d[int(i)]:
                    output_temp2.append(j+l)
            output1 = output_temp2
    return output1
```

```python
def letterCombinations_fast(digits):
    """
    :type digits: str
    :rtype: List[str]
    """
    if digits == "":
        return []
    store = [""]
    num = ["", "1", "abc", "def", "ghi", "jkl", "mno", "pqrs", "tuv", "wxyz"]
    for d in digits:
        tmpStore = []
        for val in store:
            for c in num[int(d)]:
                tmpStore.append(val + c)
        store = tmpStore
    return store
```

```python
class Solution:
    def reverseKGroup(self, head: ListNode, k: int) -> ListNode:
        if head is None or k == 1:
            return head
        
        size = self.get_size(head)
        num_iters = size // k
        
        ret = ListNode(0)
        ret.next = head
        
        prev = ret
        for _ in range(num_iters):
            prev = self.reverse(prev, k)
        return ret.next
    
    def get_size(self, head):
        curr = head
        size = 0
        while curr:
            size += 1
            curr = curr.next
        return size
    
    def reverse(self, pos, k):
        start = pos.next
        then = start.next
        for _ in range(k-1):
            start.next = then.next
            then.next = pos.next
            pos.next = then
            then = start.next
        return start
```

```python
def spiral_m(n):
    m = [[None]*n for i in range(n)]
    direction = {0:(0,1),1:(1,0),2:(0,-1),3:(-1,0)}
    count = 1
    i = 0
    j = -1
    d = 0
    while count <= n*n:
        delta_i, delta_j = direction[d]
        while i+delta_i < n and i+delta_i >=0 and j+delta_j>=0 and j+delta_j<n and m[i+delta_i][j+delta_j]==None:
            i += delta_i
            j += delta_j
            m[i][j] = count
            count += 1
        d = (d+1)%4
    return m
```

```python
for i in spiral_m(4):
    print(i)
```

# 26. Remove Duplicates from Sorted Array

```python
def removeDuplicates(nums):
    size = len(nums)
    if size <= 0:
        return 0
    new_size = 1
    j = 0
    for i in range(1, size):
        if nums[i] > nums[j]:
            j+=1
            nums[i], nums[j] = nums[j], nums[i]
            print('swap',i,j,nums)
            new_size += 1
    return new_size, nums
```
